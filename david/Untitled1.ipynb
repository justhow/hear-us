## Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/global_cia_combined.csv')
```

```python
df.shape
```

```python
len(df['country'].value_counts())
```

```python
df.head()
```

```python
df['country'].value_counts().median()
```

```python
plt.figure(figsize = (30,5))
sns.histplot(df['country'].sort_values(), )
plt.axhline(y = df['country'].value_counts().median(), color = 'orange', linewidth = 2)
plt.hlines(y = [100, 200, 300, 400, 500], xmin=0, xmax= len(df['country'].value_counts()),
           color = 'grey'
          )

plt.title('Number of Protests per Country\n', fontsize=18)
plt.xticks(rotation= 90);
```
